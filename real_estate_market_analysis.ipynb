```python
def del_locality_type(locality_name):
    locality_types = ['городской поселок ', 
                      'городской посёлок ',
                      'деревня ',
                      'коттеджный поселок ', 
                      'поселок ',
                      'поселок городского типа ',
                      'поселок станции ',
                      'посёлок ',
                      'посёлок городского типа ',
                      'посёлок при железнодорожной станции ',
                      'посёлок станции ',
                      'садоводческое некоммерческое товарищество ',
                      'садовое товарищество ',
                      'село ', 
                      'станции ', 
                      'городского типа ', 
                      'при железнодорожной станции ',
                      'коттеджный ',
                      'при железнодорожной ']
    try:
        for locality_type in locality_types:
            locality_name = locality_name.replace(locality_type, '')
        return locality_name
    except:
        return locality_name
```

```python
data['locality_name'] = data['locality_name'].apply(del_locality_type)
data['locality_name'].nunique()
```
